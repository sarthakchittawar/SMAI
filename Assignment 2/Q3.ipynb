# Imports

```python
import numpy as np
import scipy.cluster.hierarchy as hierarchy
import matplotlib.pyplot as plt
```

```python
class Hierarchial_Clustering:
    def __init__(self):
        pass
    
    def linkages(self, X, method, metric):
        return hierarchy.linkage(X, method, metric)
    
    def dendrogram(self, Z):
        return hierarchy.dendrogram(Z)
        
```

```python
f = open('SMAI-Dataset-hc-dataset/new_customers.csv', 'r')
headers = f.readline().strip().split(',')

l = []
while True:
    line = f.readline()
    if line == '':
        break
    x = line.strip().split(',')
    for i in range(len(x)):
        if i!=1:
            x[i] = int(x[i])
        elif x[i] == 'Male':
            x[i] = 0
        elif x[i] == 'Female':
            x[i] = 1
    l.append(x)
                
    
l = np.array(l)

print(headers)
print(l)
hcc =  Hierarchial_Clustering()

methods = ['single', 'complete', 'average', 'weighted']
metrics = ['euclidean']

for method in methods:
    for i in range(1, len(headers)):
        Z = hcc.linkages(l[:, :i], method, 'euclidean')
        print(method, headers[:i])
        hcc.dendrogram(Z)
        plt.show()
```

```python
f = open('SMAI-Dataset-gene-expression/gene.csv', 'r')

headers = f.readline().strip().split(',')

l = []
while True:
    line = f.readline()
    if line == '':
        break
    x = line.strip().split(',')
    x = [float(i) for i in x]
    l.append(x)

print(headers)    
print(l)
print()

l = np.array(l)

hcc =  Hierarchial_Clustering()

methods = ['single', 'complete', 'average', 'weighted']
metrics = ['euclidean']

for method in methods:
    for i in range(1, len(headers)):
        Z = hcc.linkages(l[:, :i], method, 'euclidean')
        print(method, headers[:i])
        hcc.dendrogram(Z)
        plt.show()
```

## Observations

* `Single link` - Linking two clusters based on the minimum distance between 2 elements. A drawback of this method is that it tends to produce long thin clusters since we make the link based on only 2 points.

* `Complete link` - Linking two clusters based on the max distance between 2 elements. Opposite of the problem faced with single link. Clusters tend to be overly conservative.

* `Average link` - Linkage using the distance between every pair of elements in the data.

* `Weighted link` - Linkage achieved using weighted columns in the data.

* High-dimensional data can be prone to overfitting in hierarchical clustering. 
* With more features, it becomes increasingly important to pre-process and select relevant features or apply dimensionality reduction techniques to reduce noise and computational complexity.
