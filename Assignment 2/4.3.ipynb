# Imports

```python
import numpy as np
import sklearn.mixture as mix
import matplotlib.pyplot as plt
```

```python
f = np.load('SMAI-Dataset-problem-4.3/colors.pkl', allow_pickle=True)

print(f.shape)

gmm = mix.GaussianMixture(n_components=7, covariance_type='full')
pred = gmm.fit_predict(f)

print(pred)

mean = gmm.means_
cov = gmm.covariances_

plt.scatter(f[:, 0], f[:, 1], c=pred)
plt.title('GMM Clustering')
plt.show()
```

```python
def generate_dataset(n, means, covs):
    d = np.zeros((100*n, 2))
    for i in range(n):
        d[100*i:100*(i+1), :] = np.random.multivariate_normal(means[i], covs[i], 100)
        
    return d

d = generate_dataset(7, mean, cov)

plt.scatter(d[:, 0], d[:, 1])
plt.title('Generated Dataset')
plt.show()
```

## Observations

* The above data is generated using the means, covariances of each component we estimated using GMM on our given data.
* We notice that there is a clear division of 7 clusters allocated to different colours in the first scatter plot, therefore it has been clustured into 7 different gaussians.
* When using the obtained means and covariances of the data in each cluster and a multivariate gaussian randomizer function, we obtain a dataset which is comparable to the given data.
* Therefore, we can confirm that the means, covariances and therefore GMM clustering is working upto par for this data.
